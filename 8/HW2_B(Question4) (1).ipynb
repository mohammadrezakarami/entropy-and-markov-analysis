```python
# the entropy of the Gamma distribution
import numpy as np
from scipy.special import gamma, digamma, gammaln
from scipy.integrate import quad
#Analytical entropy_func
def gamma_entropy(n):
    term1 = gammaln(n)
    term2 = -(n - 1) * digamma(n)
    term3 = n
    entropy = term1 + term2 + term3
    return entropy

def f(x, n):
    return (1 / gamma(n)) * (x ** (n - 1)) * np.exp(-x)

def log_f(x, n):
    return -np.log(gamma(n)) + (n - 1) * np.log(x) - x
#numerical_entropy_func
def numerical_integrals(n):
    entropy_integral, _ = quad(lambda x: -f(x,n) * log_f(x,n), 0, np.inf)

    return entropy_integral
#Example n:
n = 2

entropy_analytical = gamma_entropy(n)
print("Analytical entropy:", entropy_analytical)

entropy_numerical = numerical_integrals(n)
print("Numerical entropy:", entropy_numerical)
```
